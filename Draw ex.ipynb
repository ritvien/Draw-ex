```python
import turtle
#retangle
r = turtle.Turtle()
r.pensize(1)
r.pencolor("black")
r.fillcolor("green")
r.begin_fill()
#draw
r.forward(100)
r.right(90)
r.forward(75)
r.right(90)
r.forward(100)
r.right(90)
r.forward(75)
r.end_fill()
turtle.done
```
